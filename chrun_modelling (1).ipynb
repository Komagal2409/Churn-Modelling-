```python
import pandas as pd
import numpy as np
df = pd.read_csv('chrunch modeling.csv')
df.head()
```

```python
df = df.drop(['RowNumber', 'Surname', 'CustomerId'], axis='columns', errors='ignore')
df.head()
```

```python
dummies = pd.get_dummies(df[['Gender', 'Geography']], dtype=int)
dummies.head(5)
```

```python
df = pd.concat([df, dummies],axis='columns')
df.head()
```

```python
target = df['Exited']
```

```python
target.head()
```

```python
print(df.groupby('CreditScore')['Exited'].mean())
print(df.groupby('EstimatedSalary')['Exited'].mean())
print(df.groupby('Exited').mean(numeric_only=True))
```

```python
df = df.drop(['Gender','Geography','target','HasCrCard','Tenure','NumOfProducts'],axis='columns',errors='ignore')
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df, target, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
